```python
array = np.array([1,2,3,4])
```

```python
print(np.sqrt(array))
print(np.square(array))
```

```python
array2 = np.sqrt(array)
print(array2)
```

```python
scores = np.array([100,70,56,47,98,69])
```

```python
for i in scores :
    if i<70:
        print(i)
```

```python
MATCH=0
find = int(input("Find Matches Below Score: "))
for i in scores < find:
    if i == True:
        MATCH+=1
print(f"Found {MATCH} match! ")
```
